```python
import subprocess
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#to run multiple test images then run this
#note:please add the test images that you wanted to inference inside the 'data' directory
import os
type=input("Enter type")
if "src" in os.getcwd():
    os.chdir("..")
os.chdir(os.path.join(os.getcwd(),'src'))
for run_test in os.listdir("../data"):
    try:
        if type=="inference":
            subprocess.run(f"python3 main.py --img_file ../data/{run_test}",shell=True) 
            img = mpimg.imread(f"../data/{run_test}")
            plt.imshow(img)
            plt.axis()
            plt.show()
        elif type=="train":
            os.system("python3 main.py --mode train --data_dir ../image --line_mode --early_stopping 10  --fast --batch_size 500")
    except:
        subprocess.run(f"python3 main.py --img_file ../data/{run_test}",shell=True)
```
